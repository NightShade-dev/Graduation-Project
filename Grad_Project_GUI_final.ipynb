```python

import os
from google.cloud import vision_v1
import numpy as np
import io
import requests
import json
from bs4 import BeautifulSoup
from IPython.display import Image
from google.colab import drive
import pandas as pd
import csv
from IPython.display import Markdown, display
from PIL import Image
import requests
from urllib.parse import urljoin


import ipywidgets as widgets
from IPython.display import display, clear_output
import urllib.parse


from collections import namedtuple
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow
import base64

#Create GUI widgets with custom styling
query_label = widgets.Label('Enter your bag query:', style={'font-size': '24px', 'color': 'purple', 'font-weight': 'bold', 'padding': '20px'})
query_text = widgets.Text(style={'font-size': '20px', 'background-color': 'lightblue', 'border-color': 'purple', 'border-radius': '10px', 'border-width': '3px', 'padding': '15px'})
explore_button = widgets.Button(description='Explore Bags', style={'font-size': '20px', 'background-color': 'purple', 'color': 'white', 'font-weight': 'bold', 'padding': '15px'})
detect_button = widgets.Button(description='Discover from Video', style={'font-size': '20px', 'background-color': 'green', 'color': 'white', 'font-weight': 'bold', 'padding': '15px'})
clear_button = widgets.Button(description='Clear Results', style={'font-size': '20px', 'background-color': 'red', 'color': 'white', 'font-weight': 'bold', 'padding': '15px'})
output_area = widgets.Output()




#Create layout
layout = widgets.VBox([
widgets.HTML('<br><br>'),
widgets.Label('Searching about Similar', style={'font-size': '60px', 'color': 'purple', 'font-weight': 'bold'}),
widgets.HTML('<br><br>'),
detect_button,
widgets.HTML('<br><br>'),
widgets.HBox([
query_label,
query_text,
explore_button,
clear_button
], layout=widgets.Layout(justify_content='center', margin='50px')),
output_area
])

#Set alignment and background color
layout.layout.align_items = 'center'
layout.layout.background_color = '#ffeeff'

display(layout)

def query(arr):
    color_names = ['white', 'black', 'gray', 'red', 'green', 'blue', 'yellow', 'orange', 'purple', 'pink','browen']
    type_names =  ['handbag', 'backpack', 'suitcase']
    type_material =  ['leather', 'canvas', 'cuede','satin','nylon','cotton','fibre']
    object_type = ['bag','brossbag','laptopbag','shoulderbag']
    brand = ['guess','gucci','prada','Chanel','sanit laurent']

    for i in range(len(arr)):
        if arr[i].lower() in brand:
            arr[0], arr[i] = arr[i], arr[0]
        elif arr[i].lower() in color_names:
            arr[2], arr[i] = arr[i], arr[2]
        elif arr[i].lower() in type_names:
            if len(arr) > 1 and i < len(arr):
                arr[1], arr[i] = arr[i], arr[1]
        elif arr[i].lower() in type_material:
            if len(arr) > 3 and i < len(arr):
                arr[3], arr[i] = arr[i], arr[3]
        elif arr[i].lower() in object_type:
            if len(arr) > 4 and i < len(arr):
                arr[4], arr[i] = arr[i], arr[4]

    print(arr)

    # Access the value of the first key in the dictionary
    Logo_query = arr[0]
    type_object_query = arr[1] if len(arr) > 1 else ''
    color_query = arr[2] if len(arr) > 2 else ''
    matrial_query = arr[3] if len(arr) > 3 else ''
    #object_query = arr[4]

    # Encode the query variables for the search query
    name_query=type_object_query
    logo_query=Logo_query
    Color_query=color_query
    Matrial_query = matrial_query
    encoded_name_query = urllib.parse.quote(name_query)
    encoded_logo_query = urllib.parse.quote(logo_query)
    encoded_Color_query = urllib.parse.quote(Color_query)
    encoded_Matrial_query = urllib.parse.quote(Matrial_query)

    # Build the search query URL using the encoded query variables
    search_query = f'https://www.google.com/search?tbm=isch&q={encoded_name_query}+{encoded_logo_query}+{encoded_Matrial_query}+{encoded_Color_query}'

    # Print the search query URL
    print("Search Query:", search_query)

# Define event handlers
def explore_button_clicked(b):
    query_str = query_text.value
    array_query = query_str.split()

    with output_area:
        clear_output()
        query(array_query)

def detect_button_clicked(b):


  model = YOLO('yolov8x.pt')

  bag_classes = {28: 'suitcase', 26: 'handbag', 24: 'backpack'}
  bag_colors = [(255, 0, 0), (0, 255, 0), (0, 0, 255)]

  def process_video_file(file_path):
      cap = cv2.VideoCapture(file_path)
      cap.set(3, 640)
      cap.set(4, 480)

      detected_objects = set()
      object_info_set = set()

      while True:
          ret, frame = cap.read()

          if not ret:
              break

          img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
          results = model.predict(img)

          for r in results:
              boxes = r.boxes
              for i, box in enumerate(boxes):
                  c = box.cls
                  if int(c) in bag_classes:
                      # Crop the detected object
                      xmin, ymin, xmax, ymax = box.xyxy[0]
                      cropped_img = frame[int(ymin):int(ymax), int(xmin):int(xmax)]

                      # Convert cropped image to base64-encoded string
                      _, buffer = cv2.imencode('.jpg', cropped_img)
                      img_str = base64.b64encode(buffer).decode('utf-8')

                      # Compute a hash for the cropped image string
                      img_hash = hash(img_str)

                      # Check if the object is already detected
                      if img_hash in detected_objects:
                          continue

                      # Add the object hash to the set of detected objects
                      detected_objects.add(img_hash)

                      # Get the object name, color, height, and width
                      obj_name = bag_classes[int(c)]
                      obj_color = bag_colors[int(c) % len(bag_colors)]
                      obj_height = ymax - ymin
                      obj_width = xmax - xmin

                      # Create a tuple with the object information
                      object_info = {
                          'image': img_str,
                          'name': obj_name,
                          'width': obj_width,
                          'height': obj_height,
                          'color': obj_color
                      }

                      # Add the object information to the set
                      object_info_set.add(tuple(object_info.items()))

                      # Display the cropped image
                      cv2_imshow(cropped_img)

          cv2_imshow(frame)

          if cv2.waitKey(1) & 0xFF == ord('q') or len(detected_objects) >= 20:
              break

      cap.release()
      cv2.destroyAllWindows()

      return object_info_set


  file_path = '/content/Exact_bag.mp4'
  object_info_set = process_video_file(file_path)



  # Print the set with the cropped images, object names, widths, heights, and colors
  # Set the first image from object_info_set as image_query
  image_query = None
  object_name_query = None
  for obj_info in object_info_set:
      obj_info_dict = dict(obj_info)
      img_str = obj_info_dict['image']
      obj_name = obj_info_dict['name']
      obj_width = obj_info_dict['width']
      obj_height = obj_info_dict['height']
      obj_color = obj_info_dict['color']

      # Convert base64-encoded string to image
      img_data = base64.b64decode(img_str)
      nparr = np.frombuffer(img_data, np.uint8)
      cropped_img = cv2.imdecode(nparr, cv2.IMREAD_COLOR)

      print('Object:', obj_name)
      print('Width:', obj_width)
      print('Height:', obj_height)

      # Set the first image as image_query
      if image_query is None:
          image_query = cropped_img

      # Save the object name to object_name_query
      if object_name_query is None:
          object_name_query = obj_name

      cv2_imshow(cropped_img)


  # Display image_query
  cv2_imshow(image_query)

  # Print object_name_query
  print('Object Name Query:', object_name_query)



  class VisionAI:
      Label_Detection = namedtuple('Label_Detection', ('description', 'score'))
      Logo_Detection = namedtuple('Logo_Detection', ('description', 'score', 'bounding_poly'))
      Object_Detection = namedtuple('Object_Detection', ('name', 'score', 'bounding_poly'))
      Landmark_Detection = namedtuple('Landmark_Detection', ('description', 'score', 'bounding_poly', 'location'))
      Safe_Search_Detection = namedtuple('Safe_Search_Detection',
          ('adult_likelihood', 'spoof_likelihood', 'medical_likelihood', 'violence_likelihood', 'racy_likelihood'))
      Web_Detection = namedtuple('Web_Detection', ('web_entities', 'full_matching_images', 'visually_similar_images', 'pages_with_matching_images', 'best_guess_labels'))
      Web_Entity = namedtuple('Web_Entity', ('description', 'score'))
      Page_Matching_Image = namedtuple('Page_Matching_Image', ('url', 'page_title', 'full_matching_image_urls', 'partial_matching_image_urls'))
      Face_Detection = namedtuple('Face_Detection', ('detection_confidence', 'joy_likelihood', 'sorrow_likelihood', 'anger_likelihood', 'surprise_likelihood', 'under_exposed_likelihood',
          'blurred_likelihood', 'headwear_likelihood', 'bounding_poly'))

      Text_Detection = namedtuple('Text_Detection', ('description', 'bounding_poly'))

      def create_page_matching_image(url, page_title, full_matching_image_urls, partial_matching_image_urls):
          Page_Matching_Image = namedtuple('Page_Matching_Image', ('url', 'page_title', 'full_matching_image_urls', 'partial_matching_image_urls'))
          return Page_Matching_Image(url, page_title, full_matching_image_urls, partial_matching_image_urls)
      def __init__(self, client, image):
          self.client = client
          self.image = image
      #def extract_color(self):    # retutn color rgb
      #  response =self.client.image_properties(image=self.image).image_properties_annotation
      #  dominant_colors = response.dominant_colors

      #  for color in dominant_colors.colors:

       #     print('pixel fraction :{0}'.format(color.pixel_fraction))
       #     print('score: {0}'.format(color.score))
       #     print('\tred: {0}'.format(color.color.red))
       #     print('\tgreen: {0}'.format(color.color.green))
       #     print('\tblue: {0}'.format(color.color.blue))
       #    print('')

      def logo_detection(self):
          response = self.client.logo_detection(image=self.image)
          logos = response.logo_annotations
          if logos:
              results = []
              for logo in logos:
                  results.append(self.Logo_Detection(logo.description, float('{0:.2f}'.format(logo.score)), logo.bounding_poly.vertices))
              return results
          return

      def text_detection(self):
          response = self.client.text_detection(image=self.image)
          texts = response.text_annotations
          if texts:
              results = []
              for text in texts:
                  results.append(self.Text_Detection(text.description, text.bounding_poly.vertices))
              return results
          return

      def web_detection(self):
          response = self.client.web_detection(image=self.image)
          web_detection = response.web_detection
          if web_detection:
              return self.Web_Detection(
                  [self.Web_Entity(web_entity.description, float('{0:.2f}'.format(web_entity.score))) for web_entity in web_detection.web_entities],
                  [url.url for url in web_detection.full_matching_images],
                  [url.url for url in web_detection.visually_similar_images],
                  [self.Page_Matching_Image(page.url, page.page_title, [url.url for url in page.full_matching_images], [url.url for url in page.partial_matching_images]) for page in web_detection.pages_with_matching_images],
                  web_detection.best_guess_labels
              )
          return



  os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "/content/grad-project.json"
  client = vision_v1.ImageAnnotatorClient()

  # Convert the image to binary format
  _, buffer = cv2.imencode('.jpg', image_query)
  content = buffer.tobytes()
  image = types.Image(content=content)



  # Define the file path on Google Drive
  output_file_path = '/content/drive/MyDrive/output.csv'

  # Create an instance of VisionAI with your client and image
  vision_ai = VisionAI(client, image)

  # Perform logo detection
  logo_results = vision_ai.logo_detection()

  # Perform text detection
  text_results = vision_ai.text_detection()

  # Perform web detection
  web_results = vision_ai.web_detection()

  # Prepare the data for CSV
  data = []

  if logo_results:
      for logo in logo_results:
          data.append(["Logo", logo.description, logo.score, logo.bounding_poly])

  if text_results:
      for text in text_results:
          data.append(["Text", text.description, "", text.bounding_poly])

  if web_results:
      for web_entity in web_results.web_entities:
          data.append(["Web Entity", web_entity.description, web_entity.score, ""])

      for full_matching_image in web_results.full_matching_images:
          data.append(["Full Matching Image", full_matching_image, "", ""])

      for visually_similar_image in web_results.visually_similar_images:
          data.append(["Visually Similar Image", visually_similar_image, "", ""])

      for page in web_results.pages_with_matching_images:
          data.append(["Page URL", page.url, "", ""])
          data.append(["Page Title", page.page_title, "", ""])

          for full_matching_image in page.full_matching_image_urls:
             data.append(["Full Matching Image (Page)", full_matching_image, "", ""])

          for partial_matching_image in page.partial_matching_image_urls:
              data.append(["Partial Matching Image (Page)", partial_matching_image, "", ""])

      for label in web_results.best_guess_labels:
          data.append(["Best logo Label", label, "", ""])

  # Write the data to a CSV file
  with open(output_file_path, 'w', newline='') as file:
      writer = csv.writer(file)
      writer.writerows(data)

  print("Output saved to:", output_file_path)




  # Define the file path of the CSV file
  csv_file_path = '/content/drive/MyDrive/output.csv'

  # Create an empty dictionary to store the data
  data_dict = {}

  # Read the CSV file
  with open(csv_file_path, 'r') as file:
      csv_reader = csv.reader(file)

      # Iterate over each row in the CSV file
      for row in csv_reader:
          # Check if the row contains the text 'Visually Similar Image'
          if 'Visually Similar Image' in row:
              break

          # Extract the key and value from the row
          key = row[0]
          value = row[1]

          # Check if the key already exists in the dictionary
          if key in data_dict:
              # If the key exists, check if the value is a list
              if isinstance(data_dict[key], list):
                  # If it is a list, append the new value
                  data_dict[key].append(value)

              else:
                  # If it is not a list, convert the value to a list and add the new value
                  data_dict[key] = [data_dict[key], value]
          else:
              # If the key does not exist, add it to the dictionary with the value
              data_dict[key] = value

  # Print the dictionary
  for key, value in data_dict.items():
      print(key, ":", value)




  color_names = ['white', 'black', 'gray', 'red', 'green', 'blue', 'yellow', 'orange', 'purple', 'pink','browen']
  type_names =  ['handbag', 'backpack', 'suitcase']
  type_material =  ['leather', 'canvas', 'cuede','satin','nylon','cotton','fibre']
  object_type = ['bag','brossbag','laptopbag','shoulderbag']



  for i in range(len(data_dict['Web Entity'])):

      if data_dict['Web Entity'][i].lower() in color_names:
          data_dict['Web Entity'][1], data_dict['Web Entity'][i] = data_dict['Web Entity'][i], data_dict['Web Entity'][1]
      elif data_dict['Web Entity'][i].lower() in type_names:
          data_dict['Web Entity'][0], data_dict['Web Entity'][i] = data_dict['Web Entity'][i], data_dict['Web Entity'][0]
      elif data_dict['Web Entity'][i].lower() in type_material:
          data_dict['Web Entity'][2], data_dict['Web Entity'][i] = data_dict['Web Entity'][i], data_dict['Web Entity'][2]
      elif data_dict['Web Entity'][i].lower() in object_type:
          data_dict['Web Entity'][3], data_dict['Web Entity'][i] = data_dict['Web Entity'][i], data_dict['Web Entity'][3]

  print(data_dict)






  # Access the value of the first key in the dictionary
  Logo_query = data_dict.get('Logo', ' ')
  Text_query = data_dict.get('Text', [' '])
  type_object_query = data_dict.get('Web Entity', ' ')[0]
  color_query = data_dict.get('Web Entity', ' ')[1]
  matrial_query = data_dict.get('Web Entity', ' ')[2]
  object_query =data_dict.get('Web Entity', ' ')[3]
  closest_color_query = data_dict.get('Web Entity', ' ')[6]





  # Encode the query variables for the search query
  name_query=type_object_query
  logo_query=Logo_query
  Color_query=color_query
  Matrial_query = matrial_query
  encoded_name_query = urllib.parse.quote(name_query)
  encoded_logo_query = urllib.parse.quote(logo_query)
  encoded_Color_query = urllib.parse.quote(Color_query)
  encoded_Matrial_query = urllib.parse.quote(Matrial_query)

  # Build the search query URL using the encoded query variables
  search_query = f'https://www.google.com/search?tbm=isch&q={encoded_name_query}+{encoded_logo_query}+{encoded_Matrial_query}+{encoded_Color_query}'

  # Print the search query URL
  print("Search Query:", search_query)




  with output_area:
        clear_output()
        print("Search Query:", search_query)

def clear_button_clicked(b):
    with output_area:
        clear_output()

# Assign event handlers to button clicks
explore_button.on_click(explore_button_clicked)
detect_button.on_click(detect_button_clicked)
clear_button.on_click(clear_button_clicked)

# Assign event handlers to button clicks
explore_button.on_click(explore_button_clicked)
```
